### creating variables for plotting

- for 1st plot

```python
Categories = ['Category A','Category B','Category C','Category D','Category E']
Values = [23,56,41,62,19]
```

- 2nd plot

```python
Data = [12,17,21,18,14,13,16,9,12,15,19,11,14,16,20,18,15,13,16,11,10]
```

- 3rd plot

```python
Labels = ['Apple','Banana','Orange','Mango']
Sizes = [30,25,15,30]
```

- 4th plot

```python
X_values = [1,2,3,4,5]
Y_values = [2,5,3,6,9]
```

#### Importing matplotlib

```python
import matplotlib.pyplot as plt
```

### Plotting Subplots

```python
plt.subplots(ncols = 2, nrows = 2, figsize = (14,12))
plt.subplots_adjust(hspace = 0.3, wspace=0.3)
plt.subplot(2,2,1)
plt.bar(Categories, Values, color = 'r')
plt.title('Categorical Sizes')
plt.xlabel('Categories')
plt.ylabel('Sizes')

plt.subplot(2,2,2)
plt.hist(Data)
plt.title('Histogram for Data')
plt.xlabel('Data')
plt.ylabel('Occured')

plt.subplot(2,2,3)
plt.pie(Sizes, labels = Labels, autopct = '%0.1f%%')
plt.title('Fruits Quantities')
    
plt.subplot(2,2,4)
plt.scatter(X_values,Y_values, color = 'green', marker = '*')
plt.title('Scatter Plot')
plt.xlabel('X_values')
plt.ylabel('Y_values')
    
plt.show()
```
