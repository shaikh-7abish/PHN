```python
# importing panda module
import pandas as pd
```

```python
# reading csv file  
samplecsv = pd.read_csv('sample.csv')
```

```python
# converting csv file as data frame
sampleData = pd.DataFrame(samplecsv)
```

```python
# sample of data frame
sampleData.sample(10)
```

```python
# information of 1st DataFrame
sampleData.info()
```

```python
# reading excel file  
samplexlsx = pd.read_excel('sample2.xlsx')
```

```python
# converting into dataframe
sample2 = pd.DataFrame(samplexlsx)
```

```python
# sample of data frame
sample2.sample(10)
```

```python
# information of 2nd dataframe
sample2.info()
```

#merging both sample files

```python
# merging data frames into new variable
result = pd.merge(sampleData, sample2, on = 'model_name')
```

```python
# information of new merged variable
result.info()
```

```python
# removing null values 
result = result[~result['company_name'].isnull()]
```

```python
# reseting index 
result = result.reset_index(drop=True)
```

```python
# sample of our dataframr
result.sample(10)
```

```python
# information of our dataframe
result.info()
```

```python
# changing datatypes of 2 columns in our data frame
result = result.astype({'onroad_pricing':'int', 'loan_amount':'int'})
```

```python
# sample of our data frame
result.head(10)
```

```python
# gst on cars is 28%
# so adding new column with added gst value in original price
result['onroad_gst'] = ((result['onroad_pricing'] * 28 ) / 100) + result['onroad_pricing']
```
