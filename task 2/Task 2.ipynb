# Q. Find out the statistics aggregate of Age & Fare using the DataFrame.agg() method

```python
data[["Age", "Fare"]].median()
```

```python
data.agg('min')
```

```python
data.agg('max')
```

```python
data.agg('median')
```

```python
data.agg('mean')
```

```python
que3 = data.agg({"Age": ["min", "max", "median","mean"],
                 "Fare": ["min", "max", "median", "mean"]})
```
